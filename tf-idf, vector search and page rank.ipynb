```python
from collections import Counter
import math
from pymystem3 import Mystem

docs_path = 'lemmatized-texts/'
tf_idf_path = 'tf-idf/'
pagerank = 'pagerank.txt'
prs=[]


corpus=[]

links=[]

with open('text/index.txt', 'r') as f:
    i = 1
    for line in f:
        linedata = line.split('\n')
        linedata[1] = str(i) + '.txt'
        links.append(linedata) 
        i+=1

def Sort(sub_li): 
    return(sorted(sub_li, key = lambda x: float(x[2]), reverse=True))         

def lemmatize_text(text):
    return Mystem().lemmatize(text.lower())

def cosine_dic(dic1,dic2):
    numerator = 0
    dena = 0
    for key1,val1 in dic1.items():
        numerator += val1*dic2.get(key1,0.0)
        dena += val1*val1
    denb = 0
    for val2 in dic2.values():
        denb += val2*val2
    return numerator/math.sqrt(dena*denb)

for i in range(1, 101):
    with open(docs_path + str(i) + '.txt', 'r') as file:
        words = file.readline().split('\t')
        corpus.append(words)

def compute_tfidf(corpus):
    index = 0
    
    def compute_tf(text):
        tf_text = Counter(text)
        for i in tf_text:
            tf_text[i] = tf_text[i]/float(len(text))
        return tf_text
    
    def compute_idf(word, corpus):
        return math.log10(len(corpus)/sum([1.0 for i in corpus if word in i]))
    
    documents_list = []
    test_query = str(input())
    
    for text in corpus:
        tf_idf_dictionary = {}
        cosine_similarity = {}
       
        query = compute_tf(lemmatize_text(test_query))
        computed_tf = compute_tf(text)
        for word in computed_tf:
            tf_idf_dictionary[word] = computed_tf[word] * compute_idf(word, corpus)
        cosine_similarity = cosine_dic(query, tf_idf_dictionary) 
        prs.append((str(links[index][0]) + ', ' + str(links[index][1]) + ', ' +  str(cosine_similarity)).split(', '))
        index+=1
        
          #  with open(tf_idf_path + str(index) + '.txt', 'w') as file:
              #  file.write(str(tf_idf_dictionary))
       # print(str(index) + '.txt created')
       # index += 1 

compute_tfidf(corpus)
# print(pr)

with open(pagerank, 'w') as file:
    for pr in Sort(prs):
        file.write(str(pr) + '\n')
```
